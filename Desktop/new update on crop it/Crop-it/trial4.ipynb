```python
crop_data = {}

# Loop through the rows of the DataFrame
for i, row in c.iterrows():
    # Extract the values from the row
    crop_type = row['Crop Type']
    temperature = row['Temparature']
    humidity = row['Humidity ']
    nitrogen = row['Nitrogen']
    potassium = row['Potassium']
    phosphorous = row['Phosphorous']
    fertilizer_name = row['Fertilizer Name']
    
    # Check if the crop type is already in the dictionary
    if crop_type not in crop_data:
        # If it's not, add an empty dictionary for the crop type
        crop_data[crop_type] = {}
    
    # Add the data to the dictionary
    if fertilizer_name not in crop_data[crop_type]:
        crop_data[crop_type][fertilizer_name] = {
            'N': {'min': nitrogen, 'max': nitrogen},
            'P': {'min': phosphorous, 'max': phosphorous},
            'K': {'min': potassium, 'max': potassium},
            
            'humidity': {'min': humidity, 'max': humidity},
            'temperature': {'min': temperature, 'max': temperature}
        }
    else:
        crop_data[crop_type][fertilizer_name]['N']['min'] = min(crop_data[crop_type][fertilizer_name]['N']['min'], nitrogen)
        crop_data[crop_type][fertilizer_name]['N']['max'] = max(crop_data[crop_type][fertilizer_name]['N']['max'], nitrogen)
        crop_data[crop_type][fertilizer_name]['P']['min'] = min(crop_data[crop_type][fertilizer_name]['P']['min'], phosphorous)
        crop_data[crop_type][fertilizer_name]['P']['max'] = max(crop_data[crop_type][fertilizer_name]['P']['max'], phosphorous)
        crop_data[crop_type][fertilizer_name]['K']['min'] = min(crop_data[crop_type][fertilizer_name]['K']['min'], potassium)
        crop_data[crop_type][fertilizer_name]['K']['max'] = max(crop_data[crop_type][fertilizer_name]['K']['max'], potassium)
        
        crop_data[crop_type][fertilizer_name]['humidity']['min'] = min(crop_data[crop_type][fertilizer_name]['humidity']['min'], humidity)
        crop_data[crop_type][fertilizer_name]['humidity']['max'] = max(crop_data[crop_type][fertilizer_name]['humidity']['max'], humidity)
        crop_data[crop_type][fertilizer_name]['temperature']['min'] = min(crop_data[crop_type][fertilizer_name]['temperature']['min'], temperature)
        crop_data[crop_type][fertilizer_name]['temperature']['max'] = max(crop_data[crop_type][fertilizer_name]['temperature']['max'], temperature)
```
